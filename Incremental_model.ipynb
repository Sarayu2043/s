python code for incrementation using quadratic via stage method

```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_incrementation(x, a, b, c):
    """Calculate the value of the quadratic function."""
    return a * x**2 + b * x + c

# Define parameters for the quadratic function
a = 1   # Coefficient for x^2
b = 0   # Coefficient for x
c = 0   # Constant term

# Generate x values
x_values = np.linspace(0, 10, 100)  # From 0 to 10 with 100 points
# Calculate y values using the quadratic function
y_values = quadratic_incrementation(x_values, a, b, c)

# Plotting the results
plt.plot(x_values, y_values, label=f'y = {a}x^2 + {b}x + {c}')
plt.title('Quadratic Function Incrementation')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid()
plt.show()
```
